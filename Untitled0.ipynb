```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Seed for reproducibility
np.random.seed(42)

def lcg(seed, a=1664525, c=1013904223, m=2**32):
    """Linear Congruential Generator — yields uniform values in [0, 1)."""
    x = seed
    while True:
        x = (a * x + c) % m
        yield x / m

gen = lcg(42)
lcg_numbers = [next(gen) for _ in range(10)]

print("LCG Generated Uniform Numbers:")
for i, n in enumerate(lcg_numbers):
    print(f"  U({i+1}) = {n:.6f}")

mt_numbers = np.random.uniform(0, 1, 10)

print("Mersenne Twister Uniform Numbers:")
for i, n in enumerate(mt_numbers):
    print(f"  U({i+1}) = {n:.6f}")

np.random.seed(42)

samples = np.random.uniform(0, 1, 10000)

plt.figure(figsize=(8, 4))
plt.hist(samples, bins=50, color="steelblue", edgecolor="white")
plt.title("Histogram of 10,000 Uniform Random Numbers (Mersenne Twister)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("rng_histogram.png", dpi=150)
plt.show()

np.random.seed(42)

N              = 10000
mu             = 30
sigma          = 5
late_threshold = 40

delivery_times = np.random.normal(loc=mu, scale=sigma, size=N)

mean_time = np.mean(delivery_times)
std_time  = np.std(delivery_times)
min_time  = np.min(delivery_times)
max_time  = np.max(delivery_times)
p95_time  = np.percentile(delivery_times, 95)
p_late    = np.mean(delivery_times > late_threshold) * 100

print("=== Monte Carlo Simulation Results (N=10,000) ===")
print(f"Mean Delivery Time    : {mean_time:.2f} minutes")
print(f"Standard Deviation    : {std_time:.2f} minutes")
print(f"Minimum Delivery Time : {min_time:.2f} minutes")
print(f"Maximum Delivery Time : {max_time:.2f} minutes")
print(f"95th Percentile       : {p95_time:.2f} minutes")
print(f"P(Delivery > 40 min)  : {p_late:.2f}%")

plt.figure(figsize=(10, 5))
plt.hist(delivery_times, bins=60, color="steelblue",
         edgecolor="white", density=True, alpha=0.7, label="Simulated")

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 300)
plt.plot(x, stats.norm.pdf(x, mu, sigma), "r-", lw=2, label="Theoretical PDF")

plt.axvline(late_threshold, color="orange", linestyle="--",
            lw=2, label=f"Late Threshold ({late_threshold} min)")
plt.axvline(p95_time, color="green", linestyle=":", lw=2,
            label=f"95th Percentile ({p95_time:.1f} min)")

plt.title("Monte Carlo Simulation: Yellowstone Inc. Delivery Times (N=10,000)")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Probability Density")
plt.legend()
plt.tight_layout()
plt.savefig("monte_carlo_delivery.png", dpi=150)
plt.show()
```
